#### Mulicollinearity treatment

```python
# Importing necessary libraries
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Function to calculate VIF
def calc_vif(X):
    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    return vif.sort_values(by='VIF', ascending=False).reset_index(drop=True)

# Independent variables (excluding the target variable 'Heart_stroke')
independent_variables = [i for i in df1.columns if i != 'Heart_ stroke']

# Checking VIF for each variable
vif_df = calc_vif(df1[independent_variables])
vif_df
```

#### By calculating the VIF, you can identify if any of our predictor variables are highly collinear.
#### If we find variables with a high VIF (commonly above 5 or 10),
#### it may be beneficial to consider removing or combining those variables to improve the model's stability and interpretability.
#### This step is essential for ensuring that our predictive model for heart disease is reliable and not adversely affected by multicollinearity.
#### To handle multicollinearity we have used vif scores of all independdent variables which represents how well the variable is explaines by other independent variables

#### Scaling ensures that all features contribute equally to the model's decision rather than being dominated by features with larger magnitudes.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTETomek
import pandas as pd


# Define independent variables excluding the target variable
features = [col for col in dataset.columns if col != 'Heart_ stroke']

# Define the scaler and features
scaler = MinMaxScaler()

# Split data into X (features) and y (target)
X = scaler.fit_transform(dataset[features])
y = dataset['Heart_ stroke']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Handle class imbalance using SMOTE followed by Tomek links
X_smote, y_smote = SMOTETomek(random_state=42).fit_resample(X_train, y_train)

# Checking class distribution before and after handling class imbalance
for data, label in [[y_train, "Before"], [y_smote, "After"]]:
    print(f'{label} Handling Class Imbalance:')
    print(data.value_counts(), '\n')
```

```python
def analyse_model(model, X_train, X_test, y_train, y_test):
    # Fitting the model
    model.fit(X_train, y_train)

    # Evaluation and plotting for train and test sets
    for X, y, label in [(X_train, y_train, 'Train-Set'), (X_test, y_test, 'Test-Set')]:
        # Predictions and probabilities
        pred = model.predict(X)
        pred_proba = model.predict_proba(X)[:, 1]

        # Classification report
        plt.figure(figsize=(18, 5))
        plt.subplot(1, 3, 1)
        sns.heatmap(pd.DataFrame(classification_report(y_true=y, y_pred=pred, output_dict=True)).iloc[:-1, :-1].T,
                    annot=True, cmap='seismic')
        plt.title(f'{label} Classification Report')

        # Confusion matrix
        plt.subplot(1, 3, 2)
        sns.heatmap(confusion_matrix(y_true=y, y_pred=pred), annot=True, cmap='seismic', fmt='g')
        plt.title(f'{label} Confusion Matrix')
        plt.xlabel('Predicted labels')
        plt.ylabel('Actual labels')

        # AUC-ROC Curve
        plt.subplot(1, 3, 3)
        fpr, tpr, thresholds = roc_curve(y, pred_proba)
        plt.plot([0, 1], [0, 1], 'k--')
        plt.plot(fpr, tpr, label=f'AUC = {np.round(np.trapz(tpr, fpr), 3)}')
        plt.legend(loc=4)
        plt.title(f'{label} AUC-ROC Curve')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')

        plt.tight_layout()

    plt.show()

    return model
```

#### Here we are defining a function which takes classifier model and evaluates and plots for train test set and gives outputs of classification report for model performance on train and test data

#### A Support Vector Machine (SVM) is a discriminative classifier formally defined by a separating hyperplane. In other words, given labeled training data (supervised learning), the algorithm outputs an optimal hyperplane which categorizes new examples. In two dimentional space this hyperplane is a line dividing a plane in two parts where in each class lay in either side.

```python
# Compare Logistic Regression, Naive Bayes Classifier, and SVC
# Comparing ROC curves for both models
from sklearn.metrics import roc_curve, classification_report, confusion_matrix, f1_score, make_scorer, auc
def compare_models(models, model_names, X_test, y_test):
    plt.figure(figsize=(12, 8))

    for model, name in zip(models, model_names):
        model.fit(X_smote, y_smote)
        pred_proba = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, pred_proba)
        auc_score = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{name} (AUC = {auc_score:.2f})')

    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve Comparison')
    plt.legend(loc="lower right")
    plt.show()

compare_models([lr, nbc, svc], ['Logistic Regression', 'Naive Bayes Classifier', 'Support Vector Classifier'], X_test, y_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
# Define models

models = [
    ('Logistic Regression', LogisticRegression(fit_intercept=True, max_iter=10000)),
    ('Naive Bayes Classifier', GaussianNB()),
    ('Support Vector Classifier', SVC(probability=True)),
]

# Function to get evaluation metrics
def get_metrics(model, X_smote, X_test, y_smote, y_test):
    model.fit(X_smote, y_smote)
    y_pred = model.predict(X_test)
    report = classification_report(y_test, y_pred, output_dict=True)
    precision = report['1']['precision']
    recall = report['1']['recall']
    f1 = report['1']['f1-score']
    accuracy = accuracy_score(y_test, y_pred)
    return precision, recall, f1, accuracy

# Collecting metrics for each model
metrics = []
for name, model in models:
    precision, recall, f1, accuracy = get_metrics(model, X_smote, X_test, y_smote, y_test)
    metrics.append((name, precision, recall, f1, accuracy))

# Creating a DataFrame for visualization
metrics_df = pd.DataFrame(metrics, columns=['Model', 'Precision', 'Recall', 'F1 Score', 'Accuracy'])

# Plotting the metrics
plt.figure(figsize=(12, 8))
metrics_df.set_index('Model').plot(kind='bar', figsize=(14, 8), cmap='Set1')
plt.title('Model Performance Comparison', weight='bold')
plt.xlabel('Model', weight='bold')
plt.ylabel('Score', weight='bold')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

####  Accuracy: Overall, how often is the classifier correct? Accuracy = (True Pos + True Negative)/total
#### recall score: True Positive Rate: When it’s actually yes, how often does it predict yes? True Positive Rate = True Positive/actual yes
#### f1 score: The F1 score can be interpreted as a harmonic average of the precision and recall
#### Precision score: When it predicts yes, how often is it correct? Precision=True Positive/predicted yes
